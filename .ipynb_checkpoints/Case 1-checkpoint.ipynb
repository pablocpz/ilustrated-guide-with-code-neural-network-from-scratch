```python
import numpy as np
```

```python
def relu(x):
    
    return np.maximum(0, x)
```

### The weights that connects the input features to the nodes from H1 layer, will be *W_hid*, and the weights that connect & control the strenght of connection between the nodes from hidden layer H1 and the output node, will be called *W_out*

```python
def forward(X, weights):
    #every node on the hidden layer H1 takes the same input, but applies different set of weights with the same shapes
    H1_1 = relu(np.dot(X, weights["W_h1_1"]) + weights["b_h1_1"])
    H1_2 = relu(np.dot(X, weights["W_h1_2"]) + weights["b_h1_2"])
    H1_3 = relu(np.dot(X, weights["W_h1_3"]) + weights["b_h1_3"])
    

    H1 = np.concatenate([H1_1, H1_2, H1_3], axis=1)

    P = np.dot(H1, weights["W_out"]) + weights["b_out"]#multiply by the 3 weights that connect the 3 outputs of H1 and the output unit
    return P
```

For derivatives, we'll use dL_dX syntax

```python
def backward(X,Y, weights):
    """
    takes a weights dictionary & computes the gradients with respect to each one, given the error
    """
    
    M_H1 = np.dot(X, weights["W_H1"])
    
    
    N_H1 = M_H1 + weights["b_H1"]
    
    Z_H1 = relu(N_H1)
    M_out = np.dot(Z_H1, weights["W_out"]) #multiply by the 3 weights that connect the 3 outputs of H1 and the output unit
    P = M_out + weights["b_out"] #add up a single number (output's bias)
    
    
    L = (1/X.shape[0])*((Y-P)**2) #then, we calculate the Mean Squared Error Loss
    
    
    #we're ready to start computing gradients! (renember, we start derivating from outside (output) to the external functions (H1))
    dL_dP = -(2/X.shape[0])*(Y-P)
    dP_dMout = np.ones_like(M_out) #the derivative of the addend of a sum will always be one (1+2) = 3 <- changuing 1 by 2 units (3+2) =5 changues by 2
    dP_dbout = np.ones_like(weights["b_out"]) #same reasoning
    
    dMout_dZH1 = np.transpose(weights["W_out"], (1,0)) #same like d/dx (4*x) is just 4 (coefficient), but now with matrix (we nedd to transpose)
    dMout_dWout = np.transpose(Z_H1, (1,0)) #same for the another factor (since M_out = np.dot(Z_H1, W_out))
    
    #derivative of ReLU function: 0 if x<0 because it is constant here, and for x>0, f'(x) will be 1 (identity function)
    a = N_H1
    a[a<0] = 0
    a[a>0]= 1
    dZH1_dNH1 = a
    
    
    dNH1_dMH1 = np.ones_like(M_H1)
    dNH1_dbH1 = np.ones_like(weights["b_H1"])
    
    dMH1_dWH1 = np.transpose(X, (1,0)) #another multiplication, derivative are just the coefficient of the term
    


    # dL_dM_H1 = (dL_dN_H1*dN_H1_dM_H1.flatten())
    #now, let's build the final gradients
#     dL_db_out = (dL_dP*dP_dbout).sum(axis=0)
#     dL_dW_out = np.dot(dMout_dWout, dL_dMout)
#     dL_db_H1 = (dL_dN_H1*dNH1_dbH1).sum(axis=0)
    

#     dL_dW_H1 = np.dot(dMH1_dWH1, dL_dNH1)
    dL_db_out = (dL_dP*dP_dbout).sum(axis=0)
    
    dL_dMout = (dL_dP*dP_dMout)
    
    dL_dW_out = np.dot(dMout_dWout, dL_dMout)
    
    dL_dZH1 = (dL_dMout*dMout_dZH1)
    
    dL_dNH1 = (dL_dZH1*dZH1_dNH1)
    
    dL_db_H1 = (dL_dNH1*dNH1_dbH1).sum(axis=0)
    
    dL_dMH1 = (dL_dNH1*dNH1_dMH1)
    
    dL_dW_H1 = np.dot(dMH1_dWH1, dL_dMH1)
    
    
    
    # print(dL_dW_H1)
    gradients = {}
    
    gradients["b_out"] = dL_db_out
    gradients["W_out"] = dL_dW_out
    
    gradients["b_H1"] = dL_db_H1
    gradients["W_H1"] = dL_dW_H1
    
    return gradients
```

```python
def forward(X, weights):
    
    H1 = relu(np.dot(X, weights["W_H1"]) + weights["b_H1"])
    P = np.dot(H1, weights["W_out"]) + weights["b_out"]
    return P
```

```python
def train(X, Y, n_epochs, batch_size, learning_rate):
    
    def generate_batch(X, Y, start, batch_size):
        return X[start:start+batch_size], Y[start:start+batch_size]
    
    def shuffle_data(X, Y):
        permutation = np.random.permutation(X.shape[0])
        return X[permutation], Y[permutation]
    
    def init_weights():
        
        weights = {}
        
        h1_size = 3 #weights connecting H1 with output node
        
        weights["W_H1"] = np.random.randn(4,h1_size) #the 3 nodes in H1 layer, stacked together, where each one takes exactly 4 input directly from data
        weights["b_H1"] = np.random.rand(1,h1_size)
        
        weights["W_out"] = np.random.rand(h1_size, 1)
        weights["b_out"] = np.random.rand(1, 1)
        
        return weights
        
    
    start = 0
    X, Y = shuffle_data(X, Y)
    weights = init_weights()
    losses = []
    
    for i in range(n_epochs):
        if start >= X.shape[0]:
            X, Y = shuffle_data(X, Y)
            start = 0
        
        X_batch, Y_batch = generate_batch(X, Y, start, batch_size)
        
        start+=batch_size
        
        P_batch = forward(X_batch, weights)
        L_batch = (1/(X_batch.shape[0]))*(np.power(Y_batch-P_batch, 2))
        
        gradients = backward(X_batch, Y_batch, weights)
        
        
        for parameter in weights.keys():
            
            weights[parameter] -= learning_rate*gradients[parameter]
        
        losses.append(L_batch.mean())
            
        
    return weights, losses
```

```python
w, losses = train(X_train, y_train, 1000, 23, 0.001)
```
